```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
num_points = 200
a = 2.25
b = 4.24
noise = np.random.normal(0, 1, num_points)
x = np.random.uniform(0, 5, num_points)
y = a * x + b + noise
plt.scatter(x, y, color = 'b', s = 10)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
#GradientDescent
a=b=0
epochs = 1000
step = 0.0068
n = len(x)

for i in range(epochs):
    y_predicted = a*x+b
    md = (-2/n) * sum(x * (y - y_predicted))
    bd = (-2/n) * sum(y - y_predicted)
    a = a - step * md
    b = b - step * bd
print("a =", a, "b =", b)       

plt.scatter(x,y,s=10, color = 'b')
plt.plot([min(x), max(x)], [min(y_predicted), max(y_predicted)], color='y')
plt.show()
```
